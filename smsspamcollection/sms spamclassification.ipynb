# Identify the sms recieved is either **spam or ham**  

```python
#importing the libraries
import numpy as np # numerical computations
import pandas as pd # data manipulations
```

```python
# reading the smsspamcollection store into df
df=pd.read_csv('/content/SMSSpamCollection',sep='\t',names=["label","message"])
```

```python
#top 5 rows
df.head()
```

```python
# data cleaning libraries
import re # regular expression library
import nltk # text based library
nltk.download('stopwords') # downloaded the stopwords library
```

```python
# imported the libraries stopwords and stemming
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer() # initalized ps object to the porter stemmer
```

```python
# shape of the dataset
df.shape
```

```python
df.columns
```

```python
corpus=[]
for i in range(0,len(df)):
  #print()
  review=re.sub('[^a-zA-Z]'," ",df['message'][i]) # we removed all the numbers and special characters only allowed alphabets
  #print(review)
  review=review.lower() # lowering all text
  review= review.split() # splitting (converting each sentence into the list of words)
  #print("before stemming",review[1])
  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  #print("after stemming",review[1])
  review=" ".join(review)
  #print(review)
  corpus.append(review)
```

```python
# bag of words crating
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
X=cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
y=pd.get_dummies(df['label']) # converting the categorical output variable into numbers
```

```python
y=y.iloc[:,1].values
```

```python
# train and test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=.20,random_state=0)
```

```python
# training model using multinomial navie bayes algotithm
from sklearn.naive_bayes import MultinomialNB
spam_model=MultinomialNB().fit(X_train,y_train)
```

```python
#predict the model with test data
y_pred=spam_model.predict(X_test)
```

```python
# check the performance
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
y=pd.get_dummies(df['label'])
y=y.iloc[:,1].values
```

```python
# train and test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=.20,random_state=0)
```

```python
# training model using multinomial navie bayes algotithm
from sklearn.naive_bayes import MultinomialNB
spam_model=MultinomialNB().fit(X_train,y_train)
```

```python
#predict the model with test data
y_pred=spam_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
